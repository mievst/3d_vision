```python
import numpy as np
import math
from scipy.spatial.transform import Rotation as R
```

13. (5 баллов) Первая камера повернута относительно второй на 45 градусов по оси z и   
сдвинута на 10 вдоль направления оси x второй камеры. У обеих камер матрицы   
внутренних параметров единичные. Найти и распечатать фундаментальную   
матрицу.

```python
anglez = math.radians(45)
r = R.from_rotvec([0,0,anglez]).as_matrix()
r
```

```python
tx = np.array([[0, 0, 0],
               [0, 0, -10],
               [0, 10, 0]])

Tx = np.dot(tx, r)
Tx
```

```python
F = Tx @ r
F
```

Проверка решения

```python
det = np.linalg.det(F)
det
```

```python
rank = np.linalg.matrix_rank(F)
rank
```

14. (5 баллов) Первая камера повернута относительно мировой системы координат на   
45 градусов по оси z, а вторая – на -45 градусов по оси у и сдвинута на 10 в   
направлении оси x. У обеих камер матрицы внутренних параметров единичные.   
Найти и распечатать фундаментальную матрицу

```python
anglez = math.radians(-45)
angley = math.radians(-45)
r = R.from_rotvec([0,0,anglez]).as_matrix() @ R.from_rotvec([0,angley,0]).as_matrix()

tx = np.array([[0, 0, 0],
               [0, 0, -10],
               [0, 10, 0]])
Tx = np.dot(tx, r)
Tx
```

```python
F = Tx @ r
F
```

```python
det = np.linalg.det(F)
det
```

```python
rank = np.linalg.matrix_rank(F)
rank
```

15. В задаче 14 найти и распечатать оба эпиполя

Эпиполь - проекция центра камеры на изображение другой камеры

```python
anglez = math.radians(45)
P1 = np.eye(3) @ np.c_[R.from_rotvec([0,0,anglez]).as_matrix(), np.array([0, 0, 0])]
P2 = np.eye(3) @ np.c_[R.from_rotvec([0,angley,0]).as_matrix(), np.array([10, 0, 0])]
```

```python
U, S, Vt = np.linalg.svd(P1)
center1 = Vt[-1]
center1
```

```python
P1 @ center1
```

```python
U, S, Vt = np.linalg.svd(P2)
center2 = Vt[-1]
center2 = center2 / center2[3]
center2
```

```python
np.around(P2 @ center2, 3)
```

Ответ:

```python
e1 = P2 @ center1
e1
```

```python
e2 = P1 @ center2
e2
```
